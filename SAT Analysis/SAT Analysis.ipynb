```python
#Make sure extra data from the ALl row is not included

state_rate = {}
state_verbal = {}
state_math = {}

for i in sat_data:
    state_rate[i[0]] = i[1]
    state_verbal[i[0]] = i[2]
    state_math[i[0]] = i[3]
print state_rate
print state_verbal
print state_verbal
```

##### 11. Create a dictionary with the values for each of the numeric columns

```python
#Gets the data for each state
state_all = {}
for i in sat_data:
    state_all[i[0]] = i[1], i[2], i[3]
print state_all

#Creates and fills dictionaries for Participation Rate, Verbal and Math scores
rate = {}
verbal = {}
math = {}
rate_list = []
verbal_list = []
math_list = []

for i in sat_data:
    rate_list.append(i[1])
    verbal_list.append(i[2])
    math_list.append(i[3])
rate['Rate'] = rate_list
verbal['Verbal'] = verbal_list
math['Math'] = math_list

print rate
print verbal
print math
```

## Step 4: Visualize the data

##### 14. Using MatPlotLib and PyPlot, plot the distribution of the Rate using histograms.

```python
import matplotlib.pyplot as plt
%matplotlib inline
ratelist = [sat_data[i][1] for i in range(0, len(sat_data)-2)] 
plt.hist(ratelist, color = 'green')
plt.title("State's Participation Rate")
plt.xlabel("Participation Rate")
plt.ylabel("Frequency")
plt.show()
```

##### 15. Plot the Math distribution

```python
mathlist = [sat_data[i][3] for i in range(0, len(sat_data)-2)] 
plt.hist(mathlist, color = 'darkorange')
plt.title("Math SAT Scores Frequency")
plt.xlabel("Math Scores")
plt.ylabel("Frequency")
plt.show()
```

##### 16. Plot the Verbal distribution

```python
verballist = [sat_data[i][2] for i in range(0, len(sat_data)-2)] 
plt.hist(verballist, color = 'purple')
plt.title("Verbal Scores Frequency")
plt.xlabel("Verbal Scores")
plt.ylabel("Frequency")
plt.show()
```
